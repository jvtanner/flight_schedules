```python
%matplotlib inline
import flightlib
```

```python
import urllib.request
f = urllib.request.urlopen('http://web.stanford.edu/class/cs106a/flights.txt')
text = f.read().decode('utf-8')
len(text)
```

```python
flights = flightlib.parse_flights(text)
len(flights)
```

```python
import matplotlib.pyplot as plt

for city in flights.keys():
    times = []
    peoples = []
    for value in flights[city]:
        time = value[0]
        people = value[1]
        times.append(time)
        peoples.append(people)
    plt.figure(figsize=(10,3))
    plt.plot(times, peoples)
    plt.title(city)
    plt.show
```
